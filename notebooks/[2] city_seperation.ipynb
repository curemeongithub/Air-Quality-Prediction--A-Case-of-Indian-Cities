# In this notebook we seperate each city's database

```python
import numpy as np
import pandas as pd

df = pd.read_csv(r'I:\My Drive\Data Science\Sri Vipra\Code\df_cleaned_master.csv', parse_dates=['Date']) #Load the df_cleaned_master file we created in the raw_data_processing notebook

df.info()
```

```python
distinct_values = df['City'].unique()

print(distinct_values)
```

```python
# Create an empty dictionary to store the dataframes
city_dataframes = {}

# Loop through the distinct city names and create a dataframe for each city
for city in distinct_values:
    city_dataframes[city] = df[df['City'] == city]  # Use == to filter exact city names

# Now you can access the dataframe for each city using its name
print(city_dataframes['Delhi'].head())  # Example to show the dataframe for Delhi
```

```python
import os
# Directory to save the CSV files
output_dir = r"I:\My Drive\Data Science\Sri Vipra\Code\Cities Seperated"  #Enter custom folder path for depending on the environment you're working on

# Ensure the directory exists, and if not, create it
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Loop through the dictionary and save each dataframe as a CSV file
for city, city_df in city_dataframes.items():
    output_file = os.path.join(output_dir, f"{city}_data.csv")  # Create a unique filename for each city
    city_df.to_csv(output_file, index=False)  # Save without the index

    print(f"Exported {city} data to {output_file}")
```
