# In this notebook we add create lagged variables to our dataset

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\malli\Documents\GitHub\Air-Quality-Prediction--A-Case-of-Indian-Cities\data\Cities Seperated\Delhi_data.csv")

# Now let's add the (t-1) lagged variables for the relevant columns: PM2.5, PM10, NOx, NH3, CO, SO2, O3, AQI.
columns_to_lag = ['PM2.5', 'PM10', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'AQI']

# Creating lagged columns
for col in columns_to_lag:
    df[f'{col}_lag1'] = df[col].shift(1)
```
